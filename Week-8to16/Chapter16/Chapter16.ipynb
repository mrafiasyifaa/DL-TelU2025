# **Natural Language Processing with RNNs and Attention Mechanisms**

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt

print("TensorFlow version:", tf.__version__)

# ============================================================================
# 1. PERSIAPAN DATA SEDERHANA
# ============================================================================

# Contoh teks sederhana untuk demonstrasi
sample_text = """
Hello world. This is a simple text.
Natural language processing is amazing.
We are learning about RNN today.
RNN can process sequential data.
Machine learning is very powerful.
"""

print("Sample text:")
print(sample_text)
print("\nLength of text:", len(sample_text))

# ============================================================================
# 2. TOKENISASI KARAKTER
# ============================================================================

# Membuat tokenizer untuk karakter
tokenizer = keras.preprocessing.text.Tokenizer(char_level=True)
tokenizer.fit_on_texts([sample_text])

# Informasi vocabulary
vocab_size = len(tokenizer.word_index) + 1
print(f"\nVocabulary size: {vocab_size}")
print("Character to index mapping:")
for char, idx in sorted(tokenizer.word_index.items()):
    if char.isprintable():
        print(f"'{char}' -> {idx}")

# ============================================================================
# 3. PERSIAPAN DATASET UNTUK PELATIHAN
# ============================================================================

# Mengkonversi teks menjadi sequence angka
encoded_text = tokenizer.texts_to_sequences([sample_text])[0]
encoded_text = np.array(encoded_text) - 1  # Adjusting for 0-based indexing

print(f"\nEncoded text length: {len(encoded_text)}")
print(f"First 20 encoded characters: {encoded_text[:20]}")

# Membuat sequences untuk pelatihan
sequence_length = 10
sequences = []
targets = []

for i in range(len(encoded_text) - sequence_length):
    sequences.append(encoded_text[i:i + sequence_length])
    targets.append(encoded_text[i + sequence_length])

X = np.array(sequences)
y = np.array(targets)

print(f"\nTraining sequences shape: {X.shape}")
print(f"Targets shape: {y.shape}")

# ============================================================================
# 4. MEMBANGUN MODEL RNN SEDERHANA
# ============================================================================

def create_simple_rnn_model(vocab_size, embedding_dim=32, rnn_units=64):
    """
    Membuat model RNN sederhana untuk prediksi karakter
    """
    model = keras.Sequential([
        keras.layers.Embedding(vocab_size, embedding_dim,
                             input_length=sequence_length),
        keras.layers.SimpleRNN(rnn_units, return_sequences=False),
        keras.layers.Dense(vocab_size, activation='softmax')
    ])

    return model

# Membuat model
model = create_simple_rnn_model(vocab_size)

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Menampilkan arsitektur model
print("\n" + "="*50)
print("ARSITEKTUR MODEL RNN")
print("="*50)
model.summary()

# ============================================================================
# 5. PELATIHAN MODEL
# ============================================================================

print("\n" + "="*50)
print("PELATIHAN MODEL")
print("="*50)

# Callback untuk early stopping
early_stopping = keras.callbacks.EarlyStopping(
    monitor='loss',
    patience=10,
    restore_best_weights=True
)

# Melatih model
history = model.fit(
    X, y,
    epochs=50,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=1
)

# ============================================================================
# 6. VISUALISASI HASIL PELATIHAN
# ============================================================================

def plot_training_history(history):
    """
    Memplot history pelatihan
    """
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'])
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'])
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

plot_training_history(history)

# ============================================================================
# 7. FUNGSI PREDIKSI TEKS
# ============================================================================

def predict_next_char(model, tokenizer, text, temperature=1.0):
    """
    Memprediksi karakter selanjutnya berdasarkan input text
    """
    # Encode input text
    encoded = tokenizer.texts_to_sequences([text])[0]
    encoded = np.array(encoded) - 1

    # Ambil sequence terakhir
    if len(encoded) >= sequence_length:
        input_seq = encoded[-sequence_length:]
    else:
        # Pad jika terlalu pendek
        input_seq = np.pad(encoded, (sequence_length - len(encoded), 0),
                          mode='constant', constant_values=0)

    # Reshape untuk prediksi
    input_seq = input_seq.reshape(1, -1)

    # Prediksi
    predictions = model.predict(input_seq, verbose=0)
    predictions = predictions[0]

    # Apply temperature
    predictions = np.log(predictions + 1e-8) / temperature
    predictions = tf.nn.softmax(predictions).numpy()

    # Sample karakter berdasarkan probabilitas
    char_idx = np.random.choice(len(predictions), p=predictions)

    # Convert kembali ke karakter
    for char, idx in tokenizer.word_index.items():
        if idx == char_idx + 1:
            return char

    return " "  # Default jika tidak ditemukan

def generate_text(model, tokenizer, seed_text, length=50, temperature=1.0):
    """
    Generate teks dengan panjang tertentu
    """
    generated_text = seed_text

    for _ in range(length):
        next_char = predict_next_char(model, tokenizer, generated_text, temperature)
        generated_text += next_char

    return generated_text

# ============================================================================
# 8. DEMONSTRASI PREDIKSI
# ============================================================================

print("\n" + "="*50)
print("DEMONSTRASI PREDIKSI TEKS")
print("="*50)

# Test dengan berbagai seed text
seed_texts = ["Hello", "This is", "Natural", "We are"]

for seed in seed_texts:
    print(f"\nSeed: '{seed}'")

    # Generate dengan temperature berbeda
    for temp in [0.5, 1.0, 1.5]:
        generated = generate_text(model, tokenizer, seed, length=30, temperature=temp)
        print(f"Temperature {temp}: {generated}")

# ============================================================================
# 9. ANALISIS MODEL
# ============================================================================

print("\n" + "="*50)
print("ANALISIS MODEL")
print("="*50)

# Evaluasi akurasi final
final_loss, final_accuracy = model.evaluate(X, y, verbose=0)
print(f"Final Loss: {final_loss:.4f}")
print(f"Final Accuracy: {final_accuracy:.4f}")

# Statistik pelatihan
print(f"\nJumlah epoch yang dijalankan: {len(history.history['loss'])}")
print(f"Loss terbaik: {min(history.history['loss']):.4f}")
print(f"Accuracy terbaik: {max(history.history['accuracy']):.4f}")

# ============================================================================
# 10. PERBANDINGAN DENGAN MODEL YANG LEBIH KOMPLEKS
# ============================================================================

def create_gru_model(vocab_size, embedding_dim=32, rnn_units=64):
    """
    Membuat model dengan GRU untuk perbandingan
    """
    model = keras.Sequential([
        keras.layers.Embedding(vocab_size, embedding_dim,
                             input_length=sequence_length),
        keras.layers.GRU(rnn_units, return_sequences=False),
        keras.layers.Dense(vocab_size, activation='softmax')
    ])

    return model

print("\n" + "="*50)
print("PERBANDINGAN DENGAN MODEL GRU")
print("="*50)

# Membuat dan melatih model GRU
gru_model = create_gru_model(vocab_size)
gru_model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print("Training GRU model...")
gru_history = gru_model.fit(
    X, y,
    epochs=30,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=0
)

# Evaluasi perbandingan
rnn_loss, rnn_acc = model.evaluate(X, y, verbose=0)
gru_loss, gru_acc = gru_model.evaluate(X, y, verbose=0)

print(f"\nPerbandingan Performa:")
print(f"Simple RNN - Loss: {rnn_loss:.4f}, Accuracy: {rnn_acc:.4f}")
print(f"GRU        - Loss: {gru_loss:.4f}, Accuracy: {gru_acc:.4f}")

# Test generasi teks
print(f"\nPerbandingan Generasi Teks (seed: 'Hello'):")
rnn_text = generate_text(model, tokenizer, "Hello", length=40)
gru_text = generate_text(gru_model, tokenizer, "Hello", length=40)

print(f"Simple RNN: {rnn_text}")
print(f"GRU:        {gru_text}")

print("\n" + "="*50)
print("DEMONSTRASI SELESAI!")
print("="*50)
print("\nKesimpulan:")
print("1. RNN dapat mempelajari pola sekuensial dalam teks")
print("2. Model dapat memprediksi karakter berdasarkan konteks")
print("3. GRU umumnya memberikan performa yang lebih baik")
print("4. Temperature mempengaruhi kreativitas dalam generasi teks")
```
