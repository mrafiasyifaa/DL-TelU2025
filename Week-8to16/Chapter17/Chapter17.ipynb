# **Representation and Generative Learning (Autoencoders and GANs)**

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import make_blobs

# Set random seed untuk reproducibility
tf.random.set_seed(42)
np.random.seed(42)

print("TensorFlow version:", tf.__version__)

# ===== 1. DATA PREPARATION =====
print("\n=== Preparing Data ===")

# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()

# Normalize data
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0

# Take subset for faster training
X_train = X_train[:10000]
y_train = y_train[:10000]
X_test = X_test[:2000]
y_test = y_test[:2000]

print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")

# ===== 2. AUTOENCODER FOR REPRESENTATION LEARNING =====
print("\n=== Building Autoencoder ===")

# Simple Autoencoder Architecture
def build_autoencoder(input_shape, encoding_dim=32):
    # Encoder
    encoder = keras.models.Sequential([
        keras.layers.Flatten(input_shape=input_shape),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dense(64, activation='relu'),
        keras.layers.Dense(encoding_dim, activation='relu', name='bottleneck')
    ], name='encoder')

    # Decoder
    decoder = keras.models.Sequential([
        keras.layers.Dense(64, activation='relu', input_shape=[encoding_dim]),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dense(np.prod(input_shape), activation='sigmoid'),
        keras.layers.Reshape(input_shape)
    ], name='decoder')

    # Complete Autoencoder
    autoencoder = keras.models.Sequential([encoder, decoder], name='autoencoder')

    return autoencoder, encoder, decoder

# Build and compile autoencoder
autoencoder, encoder, decoder = build_autoencoder((28, 28), encoding_dim=32)
autoencoder.compile(optimizer='adam', loss='mse', metrics=['mae'])

print("Autoencoder architecture:")
autoencoder.summary()

# Train autoencoder
print("\n=== Training Autoencoder ===")
history_ae = autoencoder.fit(
    X_train, X_train,
    epochs=10,
    batch_size=256,
    validation_data=(X_test, X_test),
    verbose=1
)

# ===== 3. VISUALIZATION FUNCTIONS =====
def plot_images(images, titles=None, n_cols=5):
    n_images = len(images)
    n_rows = (n_images + n_cols - 1) // n_cols

    plt.figure(figsize=(n_cols * 2, n_rows * 2))
    for i, image in enumerate(images):
        plt.subplot(n_rows, n_cols, i + 1)
        plt.imshow(image, cmap='binary')
        plt.axis('off')
        if titles:
            plt.title(titles[i])
    plt.tight_layout()
    plt.show()

def show_reconstructions(autoencoder, X_test, n_images=10):
    # Get reconstructions
    reconstructions = autoencoder.predict(X_test[:n_images])

    # Prepare images for display
    images = []
    titles = []

    for i in range(n_images):
        images.extend([X_test[i], reconstructions[i]])
        titles.extend([f'Original {i+1}', f'Reconstructed {i+1}'])

    plot_images(images, titles, n_cols=10)

# ===== 4. REPRESENTATION LEARNING ANALYSIS =====
print("\n=== Analyzing Learned Representations ===")

# Get encoded representations
encoded_test = encoder.predict(X_test)
print(f"Original image shape: {X_test[0].shape}")
print(f"Encoded representation shape: {encoded_test[0].shape}")

# Visualize reconstructions
print("\nShowing original vs reconstructed images:")
show_reconstructions(autoencoder, X_test, n_images=5)

# t-SNE visualization of encoded representations
print("\n=== t-SNE Visualization of Learned Representations ===")
encoded_subset = encoded_test[:1000]  # Take subset for faster t-SNE
labels_subset = y_test[:1000]

tsne = TSNE(n_components=2, random_state=42, perplexity=30)
encoded_2d = tsne.fit_transform(encoded_subset)

plt.figure(figsize=(10, 8))
scatter = plt.scatter(encoded_2d[:, 0], encoded_2d[:, 1],
                     c=labels_subset, cmap='tab10', alpha=0.7)
plt.colorbar(scatter)
plt.title('t-SNE visualization of learned representations')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.show()

# ===== 5. SIMPLE GAN FOR GENERATIVE LEARNING =====
print("\n=== Building Simple GAN ===")

def build_generator(latent_dim=100):
    generator = keras.models.Sequential([
        keras.layers.Dense(128, activation='relu', input_shape=[latent_dim]),
        keras.layers.Dense(256, activation='relu'),
        keras.layers.Dense(512, activation='relu'),
        keras.layers.Dense(28*28, activation='tanh'),
        keras.layers.Reshape([28, 28])
    ], name='generator')
    return generator

def build_discriminator():
    discriminator = keras.models.Sequential([
        keras.layers.Flatten(input_shape=[28, 28]),
        keras.layers.Dense(512, activation='relu'),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(256, activation='relu'),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(1, activation='sigmoid')
    ], name='discriminator')
    return discriminator

# Build GAN components
latent_dim = 100
generator = build_generator(latent_dim)
discriminator = build_discriminator()

# Compile discriminator
discriminator.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Build and compile GAN
discriminator.trainable = False
gan = keras.models.Sequential([generator, discriminator], name='gan')
gan.compile(optimizer='adam', loss='binary_crossentropy')

print("Generator architecture:")
generator.summary()

# ===== 6. SIMPLE GAN TRAINING =====
def train_simple_gan(generator, discriminator, gan, X_train, epochs=20, batch_size=128):
    # Normalize data to [-1, 1] for tanh activation
    X_train_norm = (X_train - 0.5) * 2

    for epoch in range(epochs):
        # Train discriminator
        # Real images
        idx = np.random.randint(0, X_train_norm.shape[0], batch_size)
        real_images = X_train_norm[idx]
        real_labels = np.ones((batch_size, 1))

        # Fake images
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        fake_images = generator.predict(noise)
        fake_labels = np.zeros((batch_size, 1))

        # Train discriminator
        discriminator.trainable = True
        d_loss_real = discriminator.train_on_batch(real_images, real_labels)
        d_loss_fake = discriminator.train_on_batch(fake_images, fake_labels)
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)

        # Train generator
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        valid_labels = np.ones((batch_size, 1))

        discriminator.trainable = False
        g_loss = gan.train_on_batch(noise, valid_labels)

        # Print progress
        if epoch % 5 == 0:
            print(f"Epoch {epoch}/{epochs} - D Loss: {d_loss[0]:.4f}, D Acc: {d_loss[1]:.4f}, G Loss: {g_loss:.4f}")

print("\n=== Training Simple GAN ===")
train_simple_gan(generator, discriminator, gan, X_train, epochs=30)

# ===== 7. GENERATE NEW IMAGES =====
print("\n=== Generating New Images ===")

def generate_images(generator, n_images=10):
    noise = np.random.normal(0, 1, (n_images, latent_dim))
    generated_images = generator.predict(noise)
    # Convert from [-1, 1] to [0, 1]
    generated_images = (generated_images + 1) / 2
    return generated_images

# Generate and display new images
generated_images = generate_images(generator, n_images=10)
plot_images(generated_images, [f'Generated {i+1}' for i in range(10)])

# ===== 8. COMPARISON: REPRESENTATION VS GENERATIVE =====
print("\n=== Comparison: Representation vs Generative Learning ===")

print("\nREPRESENTATION LEARNING (Autoencoder):")
print("- Learns compact representations of input data")
print("- Encoder compresses 784-dim input to 32-dim representation")
print("- Decoder reconstructs original data from representation")
print("- Good for: dimensionality reduction, feature learning, denoising")

print("\nGENERATIVE LEARNING (GAN):")
print("- Learns to generate new data samples")
print("- Generator creates fake data from random noise")
print("- Discriminator learns to distinguish real from fake")
print("- Good for: data augmentation, creating new samples, style transfer")

# ===== 9. PRACTICAL APPLICATIONS =====
print("\n=== Practical Applications ===")

# Use autoencoder for anomaly detection
print("\nAnomalies detection using reconstruction error:")
reconstruction_errors = np.mean(np.square(X_test[:100] - autoencoder.predict(X_test[:100])), axis=(1,2))
threshold = np.percentile(reconstruction_errors, 95)

anomalies = reconstruction_errors > threshold
print(f"Found {np.sum(anomalies)} potential anomalies out of 100 samples")
print(f"Threshold: {threshold:.6f}")
print(f"Max error: {np.max(reconstruction_errors):.6f}")
print(f"Mean error: {np.mean(reconstruction_errors):.6f}")

# Show examples of high reconstruction error (potential anomalies)
high_error_indices = np.where(reconstruction_errors > threshold)[0]
if len(high_error_indices) > 0:
    print(f"\nShowing samples with high reconstruction error:")
    anomaly_samples = X_test[high_error_indices[:5]]
    anomaly_reconstructions = autoencoder.predict(anomaly_samples)

    images = []
    titles = []
    for i, idx in enumerate(high_error_indices[:5]):
        images.extend([X_test[idx], anomaly_reconstructions[i]])
        titles.extend([f'Original (Error: {reconstruction_errors[idx]:.4f})',
                      f'Reconstructed'])

    plot_images(images, titles, n_cols=10)

print("\n=== Summary ===")
print("✓ Autoencoder trained for representation learning")
print("✓ Learned 32-dimensional representations of MNIST digits")
print("✓ Visualized representations using t-SNE")
print("✓ Simple GAN trained for generative learning")
print("✓ Generated new digit-like images")
print("✓ Demonstrated anomaly detection using reconstruction error")
print("\nBoth approaches are fundamental to modern deep learning!")
```
