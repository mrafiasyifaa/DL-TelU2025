# 📎SVHN Datasets with TensorFlow

## CNN

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, roc_curve, auc
import pandas as pd
import random

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

# 1. Load and preprocess the SVHN dataset

# Load SVHN dataset
(train_ds, valid_ds, test_ds), info = tfds.load(
    'svhn_cropped',
    split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'],
    as_supervised=True,
    with_info=True,
)

# Get the number of classes
num_classes = info.features['label'].num_classes
print(f"Number of classes: {num_classes}")

# Preprocess function for each sample
def preprocess(image, label):
    # Convert image to float32 and normalize to [0, 1]
    image = tf.cast(image, tf.float32) / 255.0
    # One-hot encode the label
    label = tf.one_hot(label, num_classes)
    return image, label

# Apply preprocessing to datasets
batch_size = 128
train_ds = train_ds.map(preprocess).shuffle(10000).batch(batch_size).prefetch(tf.data.AUTOTUNE)
valid_ds = valid_ds.map(preprocess).batch(batch_size).prefetch(tf.data.AUTOTUNE)
test_ds = test_ds.map(preprocess).batch(batch_size).prefetch(tf.data.AUTOTUNE)

# Get input shape
for image_batch, label_batch in train_ds.take(1):
    input_shape = image_batch[0].shape
    print(f"Input shape: {input_shape}")

# 2. Build CNN model
def build_model(input_shape):
    model = tf.keras.Sequential([
        # First Conv Block
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.MaxPooling2D((2, 2)),

        # Second Conv Block
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        tf.keras.layers.MaxPooling2D((2, 2)),

        # Flatten and Dense layers
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.5),  # Dropout to prevent overfitting
        tf.keras.layers.Dense(num_classes, activation='softmax')
    ])

    return model

# Create the model
model = build_model(input_shape)
model.summary()

# 3. Compile and train model
# Define callbacks for early stopping and model checkpoint
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-6
)

# Compile model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
epochs = 20
history = model.fit(
    train_ds,
    validation_data=valid_ds,
    epochs=epochs,
    callbacks=[early_stop, reduce_lr]
)

# 4. Evaluate model
# Test set evaluation
test_loss, test_acc = model.evaluate(test_ds)
print(f"Test accuracy: {test_acc:.4f}")
print(f"Test loss: {test_loss:.4f}")

# Get predictions for detailed metrics
y_true = []
y_pred = []
y_pred_proba = []

for images, labels in test_ds:
    batch_preds = model.predict(images, verbose=0)
    y_pred_proba.extend(batch_preds)
    y_pred.extend(np.argmax(batch_preds, axis=1))
    y_true.extend(np.argmax(labels.numpy(), axis=1))

y_true = np.array(y_true)
y_pred = np.array(y_pred)
y_pred_proba = np.array(y_pred_proba)

# Calculate precision, recall, and F1 score
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# 5. Visualize results
def plot_training_history(history):
    # Plot training & validation accuracy values
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

def plot_roc_curves(y_true, y_pred_proba):
    # Convert to one-hot encoding for ROC curve calculation
    y_true_onehot = tf.keras.utils.to_categorical(y_true, num_classes)

    # Plot ROC curve for each class
    plt.figure(figsize=(10, 8))

    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(y_true_onehot[:, i], y_pred_proba[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'Class {i} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves for Each Digit')
    plt.legend(loc='lower right')
    plt.show()

def show_prediction_examples(ds, model, num_examples=10):
    # Get a batch from the dataset
    images, labels = next(iter(ds))

    # Make predictions
    predictions = model.predict(images)
    pred_labels = np.argmax(predictions, axis=1)
    true_labels = np.argmax(labels.numpy(), axis=1)

    # Create figure for visualization
    plt.figure(figsize=(15, 6))

    # Find some correct and incorrect predictions
    correct_idx = np.where(pred_labels == true_labels)[0]
    incorrect_idx = np.where(pred_labels != true_labels)[0]

    # Show examples of correct predictions
    plt.subplot(2, 1, 1)
    plt.title("Correct Predictions")

    for i, idx in enumerate(correct_idx[:5]):
        plt.subplot(2, 5, i+1)
        plt.imshow(images[idx].numpy())
        plt.title(f"True: {true_labels[idx]}\nPred: {pred_labels[idx]}")
        plt.axis('off')

    # Show examples of incorrect predictions
    for i, idx in enumerate(incorrect_idx[:5]):
        plt.subplot(2, 5, i+6)
        plt.imshow(images[idx].numpy())
        plt.title(f"True: {true_labels[idx]}\nPred: {pred_labels[idx]}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Generate all the visualizations
plot_training_history(history)
plot_confusion_matrix(y_true, y_pred)
plot_roc_curves(y_true, y_pred_proba)

# Show examples of predictions
show_prediction_examples(test_ds, model)

# Additional function to get per-class metrics
def print_per_class_metrics(y_true, y_pred):
    precision_per_class = precision_score(y_true, y_pred, average=None)
    recall_per_class = recall_score(y_true, y_pred, average=None)
    f1_per_class = f1_score(y_true, y_pred, average=None)

    metrics_df = pd.DataFrame({
        'Precision': precision_per_class,
        'Recall': recall_per_class,
        'F1-Score': f1_per_class
    })

    metrics_df.index.name = 'Digit Class'
    print("\nPer-class metrics:")
    print(metrics_df)

print_per_class_metrics(y_true, y_pred)

# Function to save model
def save_model(model, filename="svhn_cnn_model"):
    model.save(f"{filename}.h5")
    print(f"Model saved as {filename}.h5")

# Save the model
save_model(model)
```

## MLP

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, roc_curve, auc
import pandas as pd
import random

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)

# 1. Load and preprocess the SVHN dataset

# Load SVHN dataset
(train_ds, valid_ds, test_ds), info = tfds.load(
    'svhn_cropped',
    split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'],
    as_supervised=True,
    with_info=True,
)

# Get the number of classes
num_classes = info.features['label'].num_classes
print(f"Number of classes: {num_classes}")

# Preprocess function for each sample
def preprocess(image, label):
    # Convert image to float32 and normalize to [0, 1]
    image = tf.cast(image, tf.float32) / 255.0
    # Flatten the image for MLP input
    image = tf.reshape(image, [-1])  # Flatten the image
    # One-hot encode the label
    label = tf.one_hot(label, num_classes)
    return image, label

# Apply preprocessing to datasets
batch_size = 128
train_ds = train_ds.map(preprocess).shuffle(10000).batch(batch_size).prefetch(tf.data.AUTOTUNE)
valid_ds = valid_ds.map(preprocess).batch(batch_size).prefetch(tf.data.AUTOTUNE)
test_ds = test_ds.map(preprocess).batch(batch_size).prefetch(tf.data.AUTOTUNE)

# Get input shape
for image_batch, label_batch in train_ds.take(1):
    input_shape = image_batch[0].shape
    input_dim = image_batch[0].shape[0]  # Flattened dimension
    print(f"Input shape: {input_shape}")
    print(f"Flattened dimension: {input_dim}")

# 2. Build MLP model
def build_mlp_model(input_dim):
    model = tf.keras.Sequential([
        # Input layer
        tf.keras.layers.InputLayer(input_shape=(input_dim,)),

        # Hidden layers
        tf.keras.layers.Dense(512, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.3),

        tf.keras.layers.Dense(256, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.3),

        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dropout(0.3),

        # Output layer
        tf.keras.layers.Dense(num_classes, activation='softmax')
    ])

    return model

# Create the model
model = build_mlp_model(input_dim)
model.summary()

# 3. Compile and train model
# Define callbacks for early stopping and model checkpoint
early_stop = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=10,
    restore_best_weights=True
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-6
)

# Compile model
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
epochs = 20
history = model.fit(
    train_ds,
    validation_data=valid_ds,
    epochs=epochs,
    callbacks=[early_stop, reduce_lr]
)

# 4. Evaluate model
# Test set evaluation
test_loss, test_acc = model.evaluate(test_ds)
print(f"Test accuracy: {test_acc:.4f}")
print(f"Test loss: {test_loss:.4f}")

# Get predictions for detailed metrics
y_true = []
y_pred = []
y_pred_proba = []

for images, labels in test_ds:
    batch_preds = model.predict(images, verbose=0)
    y_pred_proba.extend(batch_preds)
    y_pred.extend(np.argmax(batch_preds, axis=1))
    y_true.extend(np.argmax(labels.numpy(), axis=1))

y_true = np.array(y_true)
y_pred = np.array(y_pred)
y_pred_proba = np.array(y_pred_proba)

# Calculate precision, recall, and F1 score
precision = precision_score(y_true, y_pred, average='weighted')
recall = recall_score(y_true, y_pred, average='weighted')
f1 = f1_score(y_true, y_pred, average='weighted')

print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

# 5. Visualize results
def plot_training_history(history):
    # Plot training & validation accuracy values
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Validation'], loc='upper left')

    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

def plot_roc_curves(y_true, y_pred_proba):
    # Convert to one-hot encoding for ROC curve calculation
    y_true_onehot = tf.keras.utils.to_categorical(y_true, num_classes)

    # Plot ROC curve for each class
    plt.figure(figsize=(10, 8))

    for i in range(num_classes):
        fpr, tpr, _ = roc_curve(y_true_onehot[:, i], y_pred_proba[:, i])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'Class {i} (AUC = {roc_auc:.2f})')

    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curves for Each Digit')
    plt.legend(loc='lower right')
    plt.show()

def show_prediction_examples(ds, model, num_examples=10):
    # Get a batch from the dataset
    images_flat, labels = next(iter(ds))

    # Make predictions
    predictions = model.predict(images_flat)
    pred_labels = np.argmax(predictions, axis=1)
    true_labels = np.argmax(labels.numpy(), axis=1)

    # Reshape flattened images back to 2D for visualization
    # Get original dimensions from the dataset info
    img_height, img_width, img_channels = info.features['image'].shape
    images = tf.reshape(images_flat, [-1, img_height, img_width, img_channels])

    # Create figure for visualization
    plt.figure(figsize=(15, 6))

    # Find some correct and incorrect predictions
    correct_idx = np.where(pred_labels == true_labels)[0]
    incorrect_idx = np.where(pred_labels != true_labels)[0]

    # Show examples of correct predictions
    plt.subplot(2, 1, 1)
    plt.title("Correct Predictions")

    for i, idx in enumerate(correct_idx[:5]):
        plt.subplot(2, 5, i+1)
        plt.imshow(images[idx].numpy())
        plt.title(f"True: {true_labels[idx]}\nPred: {pred_labels[idx]}")
        plt.axis('off')

    # Show examples of incorrect predictions
    for i, idx in enumerate(incorrect_idx[:5]):
        plt.subplot(2, 5, i+6)
        plt.imshow(images[idx].numpy())
        plt.title(f"True: {true_labels[idx]}\nPred: {pred_labels[idx]}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Generate all the visualizations
plot_training_history(history)
plot_confusion_matrix(y_true, y_pred)
plot_roc_curves(y_true, y_pred_proba)

# Show examples of predictions
show_prediction_examples(test_ds, model)

# Additional function to get per-class metrics
def print_per_class_metrics(y_true, y_pred):
    precision_per_class = precision_score(y_true, y_pred, average=None)
    recall_per_class = recall_score(y_true, y_pred, average=None)
    f1_per_class = f1_score(y_true, y_pred, average=None)

    metrics_df = pd.DataFrame({
        'Precision': precision_per_class,
        'Recall': recall_per_class,
        'F1-Score': f1_per_class
    })

    metrics_df.index.name = 'Digit Class'
    print("\nPer-class metrics:")
    print(metrics_df)

print_per_class_metrics(y_true, y_pred)

# Compare MLP vs CNN (assuming you saved the CNN results)
def compare_models(mlp_metrics, cnn_metrics=None):
    if cnn_metrics is None:
        print("No CNN metrics provided for comparison")
        return

    comparison_df = pd.DataFrame({
        'MLP': mlp_metrics,
        'CNN': cnn_metrics
    }, index=['Accuracy', 'Precision', 'Recall', 'F1-Score'])

    print("\nMLP vs CNN Comparison:")
    print(comparison_df)

    # Plot comparison
    plt.figure(figsize=(10, 6))
    comparison_df.plot(kind='bar')
    plt.title('MLP vs CNN Performance Comparison')
    plt.ylabel('Score')
    plt.ylim(0, 1.0)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Save MLP metrics for potential comparison
mlp_metrics = [test_acc, precision, recall, f1]

# Function to save model
def save_model(model, filename="svhn_mlp_model"):
    model.save(f"{filename}.h5")
    print(f"Model saved as {filename}.h5")

# Save the model
save_model(model)
```
